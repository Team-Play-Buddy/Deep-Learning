#### ì‹¤í—˜ 2. <b>digit82/kobart-summarization

```python
import requests
import torch
from transformers import PreTrainedTokenizerFast, BartForConditionalGeneration

# KoBART ìš”ì•½ ëª¨ë¸ ë¡œë“œ (ì„±ëŠ¥ ê°œì„ ëœ digit82/kobart-summarization ì‚¬ìš©)
tokenizer = PreTrainedTokenizerFast.from_pretrained("digit82/kobart-summarization")
model = BartForConditionalGeneration.from_pretrained("digit82/kobart-summarization")

# ë°°ìš° ì •ë³´ API URL (1~10ë²ˆ ë°°ìš°)
actor_ids = range(1, 11)
base_url = "http://52.78.226.136:4000/actor/{}"

# ìµœì¢… ì •ë¦¬ëœ í…ìŠ¤íŠ¸ ì €ì¥ ë¦¬ìŠ¤íŠ¸
actor_texts = []

def summarize_text(text):
    """KoBART ëª¨ë¸ì„ ì´ìš©í•´ í…ìŠ¤íŠ¸ ìš”ì•½"""
    text = text.replace("\n", " ")  # ì¤„ë°”ê¿ˆ ì œê±°
    raw_input_ids = tokenizer.encode(text, truncation=True, max_length=512)
    input_ids = [tokenizer.bos_token_id] + raw_input_ids + [tokenizer.eos_token_id]

    summary_ids = model.generate(
        torch.tensor([input_ids]),
        num_beams=4,
        max_length=512,  # ìš”ì•½ë¬¸ì˜ ìµœëŒ€ ê¸¸ì´
        eos_token_id=1  # ì¢…ë£Œ í† í°
    )

    return tokenizer.decode(summary_ids.squeeze().tolist(), skip_special_tokens=True)

for actor_id in actor_ids:
    try:
        response = requests.get(base_url.format(actor_id))

        if response.status_code == 200:
            data = response.json()

            if data.get("success") and "data" in data:
                actor = data["data"]
                profile = actor.get("actorProfile", {})
                details = actor.get("actorDetails", {})

                # ğŸ“Œ ë©”ì¸ ì •ë³´ (ë‹¨ìˆœ ë¬¸ìì—´ í•©ì¹˜ê¸°)
                main_info = f"{profile.get('nameKo', 'N/A')}, {profile.get('age', 'N/A')}ì„¸, {profile.get('height', 'N/A')}cm, {profile.get('weight', 'N/A')}kg, {'ë‚¨ì„±' if profile.get('gender') == 'male' else 'ì—¬ì„±'}"

                # ğŸ“Œ ë¶€ê°€ ì •ë³´ (ìš”ì•½ ëª¨ë¸ ì‚¬ìš©)
                extra_text = f"ê²½ë ¥: {details.get('career', 'N/A')}. íŠ¹ê¸°: {details.get('specialSkills', 'N/A')}. ìê¸°ì†Œê°œ: {details.get('bio', 'N/A')}"
                summary = summarize_text(extra_text)

                # ìµœì¢… KoCLIP ì…ë ¥ í¬ë§·
                actor_texts.append(f"ë©”ì¸ ì •ë³´: {main_info}. ë¶€ê°€ ì •ë³´: {summary}")

            else:
                actor_texts.append(f"ë°°ìš° {actor_id}: ì •ë³´ ì—†ìŒ")

        else:
            actor_texts.append(f"ë°°ìš° {actor_id}: API ìš”ì²­ ì‹¤íŒ¨ (ìƒíƒœ ì½”ë“œ {response.status_code})")

    except requests.exceptions.RequestException as e:
        actor_texts.append(f"ë°°ìš° {actor_id}: ìš”ì²­ ì¤‘ ì˜¤ë¥˜ ë°œìƒ - {e}")

# ê²°ê³¼ ì¶œë ¥
for text in actor_texts:
    print(text)
```

```python
import torch
import pandas as pd
import requests
from transformers import AutoModel, AutoProcessor

# KoCLIP ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸°
repo = "Bingsu/clip-vit-large-patch14-ko"
model = AutoModel.from_pretrained(repo)
processor = AutoProcessor.from_pretrained(repo)

BATCH_SIZE = 32  # ë°°ì¹˜ í¬ê¸°
TOTAL_ACTORS = 500  # ì´ ë°°ìš° ìˆ˜
FEATURE_DIM = 768  # KoCLIP ë²¡í„° ì°¨ì› (768ì—¬ì•¼ ì •ìƒ)

def extract_text_features(texts, actor_ids):
    """KoCLIPì„ í™œìš©í•˜ì—¬ í…ìŠ¤íŠ¸ íŠ¹ì§• ë²¡í„° ì¶”ì¶œ"""
    inputs = processor(text=texts, return_tensors="pt", padding=True)

    with torch.inference_mode():
        outputs = model.get_text_features(**inputs)

    features = outputs.cpu().tolist()

    # ë²¡í„° ì°¨ì› ê²€ì¦
    for i, feature in enumerate(features):
        if len(feature) != FEATURE_DIM:
            print(f"ğŸš¨ [ê²½ê³ ] ë°°ìš° {actor_ids[i]}: ë²¡í„° í¬ê¸° ë¶ˆì¼ì¹˜ ({len(feature)}ì°¨ì›). ìë™ ìˆ˜ì •")
            features[i] = feature[:FEATURE_DIM] + [0.0] * (FEATURE_DIM - len(feature))  # 768ì°¨ì› ë§ì¶¤

    return features

data = []
batch_texts = []
batch_actor_ids = []

for idx, actor_id in enumerate(range(1, TOTAL_ACTORS + 1), start=1):
    url = f"http://52.78.226.136:4000/actor/{actor_id}"
    response = requests.get(url)

    if response.status_code != 200:
        print(f"âš ï¸ [ì‹¤íŒ¨] ë°°ìš° {actor_id}: ì„œë²„ ì‘ë‹µ ì‹¤íŒ¨ (HTTP {response.status_code})")
        continue  # ìš”ì²­ ì‹¤íŒ¨ ì‹œ ê±´ë„ˆëœ€

    actor_data = response.json().get("data", {})
    profile = actor_data.get("actorProfile", {})

    # ë©”ì¸ ì •ë³´ ìƒì„±
    main_info = f"ë°°ìš° {profile.get('nameKo', 'ì•Œ ìˆ˜ ì—†ìŒ')}, ì„±ë³„: {profile.get('gender', 'ë¯¸ìƒ')}, ë‚˜ì´: {profile.get('age', 'ë¯¸ìƒ')}, í‚¤: {profile.get('height', 'ë¯¸ìƒ')}cm, ëª¸ë¬´ê²Œ: {profile.get('weight', 'ë¯¸ìƒ')}kg"

    batch_texts.append(main_info)
    batch_actor_ids.append(actor_id)

    # í˜„ì¬ ë°°ìš° ì²˜ë¦¬ ì™„ë£Œ ë¡œê·¸
    print(f"âœ… [ì™„ë£Œ] ë°°ìš° {actor_id}: '{main_info}' ì²˜ë¦¬ë¨")

    # ë°°ì¹˜ í¬ê¸°ë§Œí¼ ìŒ“ì´ë©´ KoCLIP ì¶”ë¡ 
    if len(batch_texts) == BATCH_SIZE:
        print(f"ğŸ”„ [ë°°ì¹˜ ì²˜ë¦¬] {BATCH_SIZE}ëª… íŠ¹ì§• ë²¡í„° ì¶”ì¶œ ì¤‘...")
        feature_vectors = extract_text_features(batch_texts, batch_actor_ids)

        for i in range(BATCH_SIZE):
            data.append([batch_actor_ids[i], batch_texts[i]] + feature_vectors[i])

        # í˜„ì¬ ë°°ì¹˜ ë²¡í„° í¬ê¸° í™•ì¸
        print(f"ğŸ“Œ [ë²¡í„° í¬ê¸° í™•ì¸] KoCLIP ë²¡í„° ì°¨ì›: {len(feature_vectors[0])} (ì •ìƒ: {FEATURE_DIM})")

        batch_texts = []
        batch_actor_ids = []

# ë‚¨ì€ ë°ì´í„° ì²˜ë¦¬
if batch_texts:
    print(f"ğŸ”„ [ë°°ì¹˜ ì²˜ë¦¬] ë‚¨ì€ {len(batch_texts)}ëª… ì²˜ë¦¬ ì¤‘...")
    feature_vectors = extract_text_features(batch_texts, batch_actor_ids)
    for i in range(len(batch_texts)):
        data.append([batch_actor_ids[i], batch_texts[i]] + feature_vectors[i])

    print(f"ğŸ“Œ [ë²¡í„° í¬ê¸° í™•ì¸] KoCLIP ë²¡í„° ì°¨ì›: {len(feature_vectors[0])} (ì •ìƒ: {FEATURE_DIM})")

# CSV ì €ì¥
columns = ["actor_id", "main_info"] + [f"feature_{i}" for i in range(FEATURE_DIM)]
df = pd.DataFrame(data, columns=columns)
df.to_csv("actor_features.csv", index=False)

print("\nâœ… ë°°ìš° 500ëª…ì˜ íŠ¹ì§• ë²¡í„°ê°€ CSVì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# CSV íŒŒì¼ ë¡œë“œ
csv_path = "actor_features.csv"
df = pd.read_csv(csv_path)

# íŠ¹ì§• ë²¡í„°ë§Œ ì¶”ì¶œ
actor_ids = df["actor_id"].tolist()
main_infos = df["main_info"].tolist()
features = df.iloc[:, 2:].values  # 768ì°¨ì› ë²¡í„° ë¶€ë¶„

# íŠ¹ì • ë°°ìš°ì™€ ìœ ì‚¬í•œ ë°°ìš° ì°¾ê¸°
def find_similar_actors(target_actor_id, top_n=3):
    if target_actor_id not in actor_ids:
        print(f"âŒ ë°°ìš° {target_actor_id} ì •ë³´ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return

    # í•´ë‹¹ ë°°ìš°ì˜ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    idx = actor_ids.index(target_actor_id)
    target_vector = features[idx].reshape(1, -1)  # 1x768 í˜•íƒœ
    target_info = main_infos[idx]

    # ëª¨ë“  ë°°ìš°ì™€ì˜ ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ê³„ì‚°
    similarities = cosine_similarity(target_vector, features)[0]

    # ë³¸ì¸ ì œì™¸í•˜ê³  ìœ ì‚¬í•œ ë°°ìš° top_n ì„ ì •
    similar_indices = np.argsort(similarities)[::-1]  # ìœ ì‚¬ë„ ë†’ì€ ìˆœ ì •ë ¬
    similar_indices = [i for i in similar_indices if actor_ids[i] != target_actor_id][:top_n]

    # ë¹„êµ ëŒ€ìƒ ë°°ìš° ì •ë³´ ì¶œë ¥
    print(f"\nğŸ­ [ìœ ì‚¬ë„ë¥¼ ë¹„êµí•  ë°°ìš°]")
    print(f"ë°°ìš° {target_actor_id} - {target_info}")

    # ê²°ê³¼ ì¶œë ¥
    print(f"\nğŸ­ [ë°°ìš° {target_actor_id}ì™€ ìœ ì‚¬í•œ ë°°ìš° TOP {top_n}]")
    for rank, i in enumerate(similar_indices, start=1):
        print(f"{rank}. ë°°ìš° {actor_ids[i]} ({main_infos[i]}) - ìœ ì‚¬ë„: {similarities[i]:.4f}")

# ë¹„êµí•  ë°°ìš° ë²ˆí˜¸ ì„¤ì •
target_actor_id = 400  # ì›í•˜ëŠ” ë°°ìš° ë²ˆí˜¸ ì…ë ¥
find_similar_actors(target_actor_id)
```
